```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
%matplotlib inline

df = pd.read_csv("master.csv")

df = df.sample(frac=1).reset_index(drop=True)
df = df.sample(n = 10000, replace = True)
```

```python
del df['country-year']
del df['HDI for year']
del df['generation']
del df['country']
del df['age']
del df['sex']

df['year'] = df['year'].astype('category')
df['year'] = df['year'].cat.codes
```

```python
df.head()
```

```python
year = df.loc[:, 'year']
print(year)
```

```python
suicides = df.loc[:, 'suicides/100k pop']
print(suicides)
```

```python
df = pd.DataFrame({"Year":year, "Suicides/100k pop":suicides})
```

```python
sbn.lineplot(x = "Year", y = "Suicides/100k pop", data=df)
```
