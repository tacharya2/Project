```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sbn
%matplotlib inline
import numpy
from sklearn.datasets import load_iris
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA
from sklearn import preprocessing

df = pd.read_csv("master.csv")
```

```python
age_groups = {'5-14 years':0, '15-24 years':1, '25-34 years':2, '35-54 years':3, '55-74 years':4, '75+ years':5}
df = df.replace({'age': age_groups})
```

```python
sbn.pairplot(df[ ['age', 'year', 'population', 'gdp_per_capita ($)', 'suicides/100k pop'] ], diag_kind = 'kde')
```

```python
import statsmodels.formula.api as sma 

m_linear_mod = sma.ols( formula = "suicides/100k pop ~ +  year + population + gdp_per_capita + age", data = df).fit() 

m_linear_mod.summary()
```
